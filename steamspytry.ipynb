```python
import requests

# Get page based on serch
S = requests.Session()

URL = "https://en.wikipedia.org/w/api.php"

SEARCHPAGE = "Skyrim video game"

PARAMS = {
    "action": "query",
    "format": "json",
    "list": "search",
    "srsearch": SEARCHPAGE
}

R = S.get(url=URL, params=PARAMS)
DATA = R.json()

PAGE = DATA['query']['search'][0]['title']
```

```python
# Get gameplay section
PARAMS = {
    "action": "parse",
    "format": "json",
    "page": PAGE,
    "prop": "sections"
}

R = S.get(url=URL, params=PARAMS)
DATA = R.json()

SECTION = 'Gameplay'
INDEX = False

for section in DATA['parse']['sections']:
    if section['line'] == SECTION:
        INDEX = section['index']
        break
```

```python
# Get section wikitext
PARAMS = {
    "action": "parse",
    "format": "json",
    "page": PAGE,
    "section": INDEX,
    "prop": "wikitext"
}

R = S.get(url=URL, params=PARAMS)
DATA = R.json()

wikicode = DATA['parse']['wikitext']['*']
```

```python
import mwparserfromhell

# Parse wikicode
parsed_wikicode = mwparserfromhell.parse(wikicode).strip_code()
parsed_wikicode
```
